```python
#1. What is the average price of all cars in the dataset?

df['Price'].mean()
```

```python
# 2. Which car brand has the highest average price?

#df.groupby("Brand")["Price"].mean().idxmax()

df.groupby("Brand")["Price"].mean().sort_values(ascending=False).head()
```

```python
#3. What is the most common fuel type in the dataset?

# df['Fuel_Type'].mode()

df['Fuel_Type'].value_counts().idxmax()
```

```python
#4. How many cars have a mileage greater than 200,000?

len(df[df['Mileage'] > 200000])
```

```python
# 6. Which car model has the highest price?

df['Price'].max()
```

```python
df[df['Price'] == df['Price'].max() ]['Model']
```

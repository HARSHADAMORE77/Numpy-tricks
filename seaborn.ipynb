```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the tips dataset
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
tips
```

```python
# Scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
# Line plot
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title("Total Bill vs Tip (Line Plot)")
plt.show()
```

```python
# Bar plot
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day")
plt.show()
```

```python
# Histogram
sns.histplot(tips["total_bill"], kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# Box plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day (Box Plot)")
plt.show()
```

```python
# Violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day (Violin Plot)")
plt.show()
```

```python
# Pair plot
sns.pairplot(tips, hue="sex")
plt.show()
```

```python

# Facet grid
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

```python
# Customizing a scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time", style="smoker", size="size")
plt.title("Customized Scatter Plot")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend(title="Time", loc="upper left")
plt.show()
```

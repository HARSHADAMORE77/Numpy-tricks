```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot 1
ax1.plot(x, y1, color='blue')
ax1.set_title('Sine Wave')

# Plot 2
ax2.plot(x, y2, color='red')
ax2.set_title('Cosine Wave')

# Add common labels
fig.suptitle('Multiple Plots Example')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Mask large values of tan(x) to avoid vertical asymptotes in the plot
y3 = np.ma.masked_where(np.abs(y3) > 10, y3)

# Create subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 4))

# Plot 1: Sine
ax1.plot(x, y1, color='blue')
ax1.set_title('Sine Wave')

# Plot 2: Cosine
ax2.plot(x, y2, color='red')
ax2.set_title('Cosine Wave')

# Plot 3: Tangent
ax3.plot(x, y3, color='green')
ax3.set_title('Tangent Wave')
ax3.set_ylim(-10, 10)  # Limit y-axis for better visualization

# Add common title
fig.suptitle('Multiple Plots Example')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a plot with customizations
plt.plot(x, y, marker='o', linestyle='--', color='purple', linewidth=2, markersize=8, label='Custom Line')

# Add grid, labels, and title
plt.grid(True, linestyle=':', alpha=0.7)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.title('Customized Plot', fontsize=14)
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')



plt.savefig("3dplot222.png",dpi = 300)

# Show the plot
plt.show()
```
